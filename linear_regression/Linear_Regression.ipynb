```python
#Nicolás Hernández Adair 7BV1
import pandas as pd
import numpy as np
import tkinter as tk
from tkinter import filedialog, messagebox
import matplotlib.pyplot as plt
import seaborn as sns

from pyspark.sql import SparkSession
from pyspark.ml import Pipeline
from pyspark.ml.feature import (
    StandardScaler,
    StringIndexer,
    OneHotEncoder,
    VectorAssembler,
    Interaction,
)
from pyspark.ml.regression import LinearRegression
from pyspark.ml.linalg import Vectors
from pyspark.sql.functions import col, when
from pyspark.sql.types import IntegerType
from pyspark.sql.functions import pandas_udf

# Global variables
ROUTE = ""
SELECTED_COLUMNS = ["age", "bmi", "children", "charges"]
CATEGORICAL_COLUMNS = ["smoker", "sex", "region"]
# GUI functions
def open_file():
    global ROUTE
    file_path = filedialog.askopenfilename()

    if file_path == "":
        open_file()
    route_label.config(text=file_path)


def get_direction():
    global ROUTE
    if len(route_label["text"]) == 0:
        messagebox.showinfo(title="Error", message="Make sure to enter a path.")
        return
    ROUTE = route_label["text"]
    window.destroy()


window = tk.Tk()
window.config(pady=30, padx=30)
window.title("Regression")

# Buttons
open_button = tk.Button(window, text="Select File", command=open_file)
open_button.grid(column=0, row=0)

start_button = tk.Button(window, text="Load", command=get_direction)
start_button.grid(column=0, row=1, columnspan=2)

# Labels
route_label = tk.Label(window, text="")
route_label.grid(column=1, row=0)

window.mainloop()

# Initialize Spark
spark = SparkSession.builder.appName("Regression").getOrCreate()

# Data processing functions
def open_archive(file_path):
    df = spark.read.csv(file_path, header=True, inferSchema=True)
    return df

def summarize_dataframe(df):
    print("\n\nDataFrame Data")
    df.show(5)

    print("\n\nStatistical Information")
    df.describe().show()

    print("\n\nNull Values Per Column")
    for column in df.columns:
        null_count = df.filter(col(column).isNull()).count()
        print(f"Column '{column}': {null_count} null values")

def get_histogram(df, column):
    feature_values = df.select(column).rdd.flatMap(lambda x: x).collect()
    
    plt.hist(feature_values, bins=20)
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.title(f"{column} Histogram")
    plt.show()

def informative_feature_correlation(df, columns):
    df = df.toPandas()
    sns.set(style="ticks")

    sns.pairplot(df[columns], diag_kind="kde", markers="o")

    plt.show()

def convert_categorical_to_dummies(df, categorical_columns):
    indexers = [StringIndexer(inputCol=col, outputCol=col + "_index") for col in categorical_columns]
    encoders = [OneHotEncoder(inputCol=col + "_index", outputCol=col + "_onehot") for col in categorical_columns]
    stages = indexers + encoders
    pipeline = Pipeline(stages=stages)
    model = pipeline.fit(df)
    transformed_df = model.transform(df)

    for col in categorical_columns:
        transformed_df = transformed_df.drop(col)

    return transformed_df

def generate_model(df):
    assembler = VectorAssembler(inputCols=df.columns, outputCol="features")
    assembled_df = assembler.transform(df)

    lr = LinearRegression(featuresCol="features", labelCol="charges")
    model = lr.fit(assembled_df)

    coefficients = model.coefficients
    intercept = model.intercept

    return model, intercept, coefficients, assembled_df

def non_linear(value):
    return value**2

def improve_model(df, assembled_df):
    non_linear_udf = pandas_udf(non_linear, IntegerType())
    insurance_improve = df.withColumn('age2', non_linear_udf(df['age']))
    
    insurance_improve = insurance_improve.withColumn("bmi30", when(col("bmi") >= 30, 1).otherwise(0))
    
    interaction = Interaction(inputCols=['bmi30', 'smoker_onehot'], outputCol="bmi30:smoker_ohe")
    insurance_improve = interaction.transform(insurance_improve)
    
    columns2 = insurance_improve.columns
    columns2.pop(3)
    
    vector_assembler_improved = VectorAssembler(inputCols=columns2, outputCol="features_improve")
    assembled_improved_df= vector_assembler_improved.transform(insurance_improve)
    
    data_improve = assembled_improved_df.select('features_improve', 'charges')
    
    train_data, test_data = data_improve.randomSplit([0.75, 0.25])
    model2 = LinearRegression(featuresCol='features_improve', labelCol='charges')
    model2 = model2.fit(train_data)
    return model2

def evaluate_model(lr_model):
    summary = lr_model.summary

    print("Degrees of Freedom: " + str(summary.degreesOfFreedom))
    print("Deviance Residuals: " + str(summary.devianceResiduals))
    print("Explained Variance: " + str(summary.explainedVariance))
    print("Mean Absolute Error: " + str(summary.meanAbsoluteError))
    print("Mean Squared Error: " + str(summary.meanSquaredError))
    print("Num Instances: " + str(summary.numInstances))
    print("Root Mean Squared Error: " + str(summary.rootMeanSquaredError))
    print("R-squared: " + str(summary.r2))

# Open file and preprocess data
df = open_archive(ROUTE)
df = convert_categorical_to_dummies(df, CATEGORICAL_COLUMNS)
df.show()

# Data summary
summarize_dataframe(df)

# Correlation visualization
informative_feature_correlation(df, SELECTED_COLUMNS)

# Histograms
for i in range(0, len(SELECTED_COLUMNS)):
    get_histogram(df, SELECTED_COLUMNS[i])

# Generate and evaluate the initial model
model, intercept, coefficients, assembled_df = generate_model(df)
print("Coefficients and Intercept\n")
print("Coefficients:", coefficients)
print("Intercept:", intercept)
print("\n\nEvaluating the model\n")
evaluate_model(model)

# Improve the model
model2 = improve_model(df, assembled_df)
print("\n\nEvaluating the model improved\n")
evaluate_model(model2)
```
